```python
from ultralytics import YOLO
from pathlib import Path
import cv2
import os

# Load trained model
model = YOLO(
    r"C:\Users\Priyanka\Downloads\Wheel_Center_Notch_Detection.v1i.yolov8\yolo11_results\weights\best.pt"
)

# Input folder
input_root = r"C:\Users\Priyanka\Downloads\Wheel_Center_Notch_Detection.v1i.yolov8\CLEAN_RGB"

# Output folder
output_root = r"C:\Users\Priyanka\Downloads\Wheel_Center_Notch_Detection.v1i.yolov8\PREDICTED_IMAGES_YOLO11"

os.makedirs(output_root, exist_ok=True)

total_images = 0

# Find all images
all_images = list(Path(input_root).rglob("*.png"))
all_images += list(Path(input_root).rglob("*.jpg"))
all_images += list(Path(input_root).rglob("*.jpeg"))

print(f"Found {len(all_images)} images")

for img_path in all_images:

    img = cv2.imread(str(img_path))

    results = model.predict(
        str(img_path),
        conf=0.25,
        verbose=False
    )

    for r in results:

        for box in r.boxes:

            cls = int(box.cls[0])

            x1, y1, x2, y2 = map(int, box.xyxy[0])

            label = model.names[cls]

            # Different colors
            if label == "center":
                color = (0, 255, 0)      # Green
            else:
                color = (0, 0, 255)      # Red

            # Draw bounding box
            cv2.rectangle(
                img,
                (x1, y1),
                (x2, y2),
                color,
                2
            )

            # Display coordinates
            text = f"{label}: ({x1},{y1}) ({x2},{y2})"

            cv2.putText(
                img,
                text,
                (x1, max(20, y1 - 10)),
                cv2.FONT_HERSHEY_SIMPLEX,
                0.5,
                color,
                2
            )

    # Preserve folder structure
    relative_path = img_path.relative_to(input_root)
    save_path = Path(output_root) / relative_path

    save_path.parent.mkdir(parents=True, exist_ok=True)

    cv2.imwrite(str(save_path), img)

    total_images += 1

    if total_images % 20 == 0:
        print(f"Processed {total_images} images")

print("\nDone!")
print(f"Processed {total_images} images")
print(f"Results saved in:\n{output_root}")
```
